# Задание 2.2
Реализовать регрессию вручную, отобразить наклон, сдвиг и MSE.

```python
X = df[X_feature].values.astype(float)
y = df[y_target].values.astype(float)

# функция среднеквадратичной ошибки
def mserror(w0, w1, X, y):
    y_pred = w0 + w1 * X
    return np.mean((y - y_pred) ** 2)

# функция градиента
def gr_mserror(w0, w1, X, y):
    y_pred = w0 + w1 * X
    error = y_pred - y

    grad_w0 = 2 * np.mean(error)
    grad_w1 = 2 * np.mean(error * X)

    return grad_w0, grad_w1

w0 = 0.0
w1 = 0.0
learning_rate = 0.01
eps = 1e-6
max_iter = 100000

for i in range(max_iter):
    grad_w0, grad_w1 = gr_mserror(w0, w1, X, y)
    new_w0 = w0 - learning_rate * grad_w0
    new_w1 = w1 - learning_rate * grad_w1
    if abs(new_w0 - w0) <= eps and abs(new_w1 - w1) <= eps:
        w0, w1 = new_w0, new_w1
        break
    w0, w1 = new_w0, new_w1

mse_final = mserror(w0, w1, X, y)

print("Результаты линейной регрессии (градиентный спуск):")
print(f"Наклон (w1): {w1:.6f}")
print(f"Сдвиг (w0): {w0:.6f}")
print(f"MSE: {mse_final:.6f}")
print(f"Число итераций: {i + 1}")
```

# Задание 3.1
Выполнить однофакторный ANOVA тест, чтобы проверить влияние
региона на индекс массы тела (BMI), используя первый способ,
через библиотеку Scipy

```python
!pip install scipy
```

```python
from scipy.stats import f_oneway

groups = []
for r in df_insurance["region"].unique():
    groups.append(df_insurance[df_insurance["region"] == r]["bmi"])

f_stat, p_value = f_oneway(*groups)

print("ANOVA через SciPy:")
print("F-статистика:", f_stat)
print("p-value:", p_value)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols("bmi ~ C(region)", data=df_insurance).fit()
anova_table = anova_lm(model)

anova_table
```

# Задание 3.3
С помощью t критерия Стьюдента перебрать все пары. Определить
поправку Бонферрони. Сделать выводы.

```python
from scipy.stats import ttest_ind
import itertools

regions = df_insurance["region"].unique()
pairs = list(itertools.combinations(regions, 2))

alpha = 0.05
m = len(pairs)
bonf_alpha = alpha / m

print(f"Поправка Бонферрони: новое критическое значение alpha = {bonf_alpha:.4f}\n")

for a, b in pairs:
    group_a = df_insurance[df_insurance["region"] == a]["bmi"]
    group_b = df_insurance[df_insurance["region"] == b]["bmi"]
    stat, p = ttest_ind(group_a, group_b)

    if p < bonf_alpha:
        conclusion = "Гипотеза отклоняется (есть различия)"
    else:
        conclusion = "Гипотеза принимается (различий нет)"

    print(f"{a} vs {b}: p = {p:.5f} - {conclusion}")
```
